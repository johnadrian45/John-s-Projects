```
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
import statsmodels.api as sm

postgres_user = 'dsbc_student'
postgres_pw = '7*.8G9QH21'
postgres_host = '142.93.121.174'
postgres_port = '5432'
postgres_db = 'weatherinszeged'

engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))
weather_data = pd.read_sql_query('select * from weatherinszeged', con = engine)
engine.dispose

weather_data.info()
weather_data.head()
```

```
weather_data['target'] = weather_data.apparenttemperature - weather_data.temperature
Y = weather_data.target
X = weather_data[['humidity', 'windspeed']]

X = sm.add_constant(X)
results = sm.OLS(Y,X).fit()
print(results.summary())

#it would seem that all all the coefficients are statistically significant although
# i would expect that the greater the humidity the greater the difference between
# the apparent temperature and temperature. Also I would expect the same for
# windspeed, but this seems to not be true for either case.
```

```
weather_data['humidity*windspeed'] = weather_data.humidity * weather_data.windspeed
X = weather_data[['humidity', 'windspeed', 'humidity*windspeed']]
X = sm.add_constant(X)
results = sm.OLS(Y,X).fit()
print(results.summary())

# Most of the coefficients changed, it also seems that humidity and windspeed
# now have a positive correlation while humidity*windspeed is negative. All our
# p values are within acceptable margins. This essentially means that the 
# effects of windspeed are mitigated by windspeed
```

```
postgres_user = 'dsbc_student'
postgres_pw = '7*.8G9QH21'
postgres_host = '142.93.121.174'
postgres_port = '5432'
postgres_db = 'houseprices'

engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))

houseprices_data = pd.read_sql_query('select * from houseprices', con = engine)
engine.dispose()

houseprices_data.info()
houseprices_data.head()
```

```
X = houseprices_data[['overallqual', 'grlivarea', 'garagecars', 'garagearea', 'totalbsmtsf']]
Y = houseprices_data.saleprice
X = sm.add_constant(X)
results = sm.OLS(Y,X).fit()
print(results.summary())

# It seems that there are generally positive trends with all the features and the
# target and all the targer have p values within the acceptable range. 
```

```
houseprices_data['garagecars*garagearea'] = houseprices_data.garagecars * houseprices_data.garagearea
X = houseprices_data[['overallqual', 'grlivarea', 'garagecars', 'garagearea', 'totalbsmtsf', 'garagecars*garagearea']]
X = sm.add_constant(X)
results = sm.OLS(Y,X).fit()
print(results.summary())

# After creating a feature that combines the number of cars in a garage and the
# garage square footage, we see that  the pvalue of garage cars is now .29 meaning
# that it is a useless feature. For a given garage area if we raise the number
# of cars by one, the price would change by  3476 + 35*garagearea
```
