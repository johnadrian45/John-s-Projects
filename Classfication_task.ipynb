```
data = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/00468/online_shoppers_intention.csv')
data.info()

data.nunique()
```

```
# What types are its attibutes?
dtypes = []
for col in data:
  dtypes.append(data[col].dtypes)
dtypes
```

```
# Is there any missing data?
continuous_variables = []
categorical_variables = []
for col in data:
  if (data[col].dtypes != 'object') & (data[col].dtypes != 'bool'):
    continuous_variables.append(col)
  else:
    categorical_variables.append(col)

# See if there are any missing values
for col in categorical_variables:
  print('\nUnique vales for {}: {}'.format(col, data[col].unique()))

for col in continuous_variables:
  ratio = data[col].count()/len(data[col])
  print('Missing values ration for {}: {}'.format(col, ratio))

#There are no missing values
```
