```
import pandas as pd

census_population_df = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/state-populations.csv')
census_regions_df = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/census-divisions.csv')

census_df = pd.merge(census_population_df, census_regions_df)
census_df.head()

```

```
census_pivot = census_df.pivot(index = '2010',columns= 'region', values='state')
census_pivot.head().reset_index()
```
