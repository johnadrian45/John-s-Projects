```python
def find_avg(data):
    total_sum = 0
    for number in data:
        total_sum += number
    count = len(data)
    avg = total_sum/count
    return avg
    
    
query_3 = '''
SELECT
    length
FROM
    film
'''
movie_lengths = engine.execute(query_3)
rows = movie_lengths.fetchall()
lengths = [row['length'] for row in rows]
find_avg(lengths)
```

```python
import math

def find_median(data):
    data.sort()
    data_length = len(data)
    if data_length % 2 ==  0:
        middle_1 = math.floor(data_length/2)
        middle_2 = middle_1 - 1
        median = (data[middle_1] + data[middle_2])/2
    else:
        median = data[math.floor(data_length/2)]
    return median
find_median(lengths)
```

```python
def find_std(data):
    x_bar = find_avg(data)
    N = len(data)
    total = 0
    for number in data:
        diff = number - x_bar
        squared = math.pow(diff, 2)
        total += squared
    sigma = math.sqrt(total/(N-1))
    return sigma
find_std(lengths)
```
