```
import requests 

#the url variable is constant as it contains the url and access key
url = "http://api.harvardartmuseums.org/person?apikey=e34cb650-6bcb-11ea-b5c2-0b58570f384c"

# data is returned in pages of 10 entries, and the index refers to which entry
# is being called. query is a constant that refers to the culture being called
# and which page of the data returned is being accesed.
page = 1
index = 0
query = {'q' : 'culture:British', 'page':page}

# method that requests data
def request_data(url, query):
  response = requests.get(url, query)
  if response.ok:
    data = response.json()
  return data

# method that extracts the meta data about the data returned by the API
def extract_meta_data(data):
  meta_data = data['info']
  return meta_data

# method that extracts record data returned by the API
def extract_records(data):
  records = data['records']
  return records

# method that extracts names kept in the record
def extract_name(records, index):
  name = records[index]['alphasort']
  return name

# method that extracts birthplaces kept in the record
def extract_birthplace(records, index):
  birthplace = records[index]['birthplace']
  return birthplace
    

data = request_data(url, query)

meta_data = extract_meta_data(data)
total_pages = meta_data['pages']
records = extract_records(data)
total_pages = extract_meta_data(request_data(url,query))['pages']
names = {}

while page <= total_pages:
  data = request_data(url, query)
  records = extract_records(data)
  while index < len(records):
    names[extract_name(records,index)] = extract_birthplace(records, index)
    index = index + 1
  page = page + 1
  index = 0
  query = query = {'q' : 'culture:British', 'page':page}

names
data


```
