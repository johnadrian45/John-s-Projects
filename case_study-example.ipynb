```
PATH = 'https://raw.githubusercontent.com/Thinkful-Ed/data-201-resources/master/1.4.11_experimentation_guided_example/experiment_data_2_wks.csv'
experiment = pd.read_csv(PATH)

#How large is each group?
test = experiment['test'] == 1
control = experiment['test'] == 0
experiment_test = experiment.iloc[test.values]
experiment_control = experiment.iloc[control.values]

print(experiment_test.shape[0])
print(experiment_control.shape[0])
```

```
#What is the signup rate for test and control

test_sign_up = experiment_test['is_signed_up'] == 1
experiment_test_signup = experiment_test.iloc[test_sign_up.values]
test_rate = experiment_test_signup.shape[0]/experiment_test.shape[0]



control_sign_up = experiment_control['is_signed_up'] == 1
experiment_control_signup = experiment_control.iloc[control_sign_up.values]
control_rate = experiment_control_signup.shape[0]/experiment_control.shape[0]

print(test_rate)
print(control_rate)


```
